## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x * y * z

<img src="https://raw.githubusercontent.com/gjhernandezp/NM/main/SymPyLab/sympylabwolfram7.jpg" />

````
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)
x = x1 + I*x2
y = y1 + I*y2
z = z1 + I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
````

# **My own example Symbolic and Numerical Complex Evaluations**



```
# x = x1**3 + I*x2/2,y = 4*y1 + I*y2**2, z = 6*z1 + I*z2, x * y / z
```



image.png

````
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)
x = x1**3 + I*x2/2
y = 4*y1 + I*y2**2
z = 6*z1 + I*z2

print(x*y/z)
print(expand(x*y/z))
print(expand((x*y)/z))
print(expand(x*(y/z)))
w=(2*pi)/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
````
