# Modificaciones
- Algoritmo de Machine Learning: Se cambió la regresión lineal por un Regresor de Bosque Aleatorio. De esta manera, combina múltiples árboles de decisión para capturar relaciones complejas y no lineales en los datos, mientras que el regresor lineal asume que la relación entre las características y la variable objetivo es estrictamente lineal. Esta capacidad de modelar interacciones y patrones no lineales permite al bosque aleatorio manejar conjuntos de datos con alta dimensionalidad, ruido, o características correlacionadas de manera más efectiva, reduciendo el riesgo de subajuste y mejorando la precisión de las predicciones en problemas más complejos.
